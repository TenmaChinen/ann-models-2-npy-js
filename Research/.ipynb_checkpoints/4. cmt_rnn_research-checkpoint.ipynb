### LSTM Statefull Class

```python
sigmoid = lambda x : 1 / ( 1 + np.exp(-x ))
softsign = lambda x : x / ( 1 + abs(x))

class LSTM:
    def __init__(self,units,kernel,bias,rec_kernel):
        self.units = units
        self.w_i, self.w_f, self.w_c, self.w_o = np.split(kernel, 4 , axis = 1 )
        self.b_i, self.b_f, self.b_c, self.b_o = np.split(bias, 4 , axis = 0 )
        self.u_i, self.u_f, self.u_c, self.u_o = np.split(rec_kernel, 4 , axis = 1 )
                 
    def __call__(self,x):
        
        l_states = []
        for x_batch in x:
            
            h_t_1 = np.zeros((1,self.units)) # h_{t-1}
            c_t_1 = np.zeros((1,self.units)) # c_{t-1}

            for x_t in x_batch[:,None]:
                i_t = sigmoid( np.dot(x_t, w_i) + np.dot(h_t_1, u_i) + b_i )

                c_t_tilda = softsign(np.dot(x_t, w_c) + np.dot(h_t_1, u_c) + b_c) # tanh

                f_t = sigmoid( np.dot( x_t, w_f ) + np.dot( h_t_1, u_f ) + b_f )

                c_t = i_t * c_t_tilda + f_t * c_t_1
                o_t = sigmoid( np.dot( x_t, w_o ) + np.dot( h_t_1, u_o ) + b_o )
                h_t = o_t * softsign( c_t ) # tanh

                h_t_1 = h_t
                c_t_1 = c_t
            
            l_states.append(h_t)

        return np.vstack(l_states)

units = l_layers_cnf[0]['units'] # 20
kernel = lstm_cell['kernel:0'][:]
bias = lstm_cell['bias:0'][:]
rec_kernel = lstm_cell['recurrent_kernel:0'][:]

lstm = LSTM(units,kernel,bias,rec_kernel)
states = lstm(x = x_test )
y_preds = states.dot(w_out) + b_out

# print(states.shape)
fig , ax = subplots( figsize = ( 15 , 5 ) )
ax.plot(r_test,y_test, lw = 10, label = 'Test')
ax.plot(r_test,y_preds, lw = 5, ls = '--', label = 'Prediction')
ax.legend();
```

### STATEFUL

```python
class LSTM:
    def __init__(self,units):
        self.units = units
        self.h_t_1 = np.zeros((1,units)) # h_{t-1}
        self.c_t_1 = np.zeros((1,units)) # c_{t-1}

    def reset_states(self):
        self.h_t_1 = np.zeros((1,self.units)) # h_{t-1}
        self.c_t_1 = np.zeros((1,self.units)) # c_{t-1}
        
    def warm_up(self,x):
        
        self.reset_states()
        h_t_1 , c_t_1 = self.h_t_1, self.c_t_1
        
        for x_t in x:
            x_t = x_t.reshape(1,1)
            i_t = sigmoid( np.dot(x_t, w_i) + np.dot(h_t_1, u_i) + b_i )
    
            c_t_tilda = softsign(np.dot(x_t, w_c) + np.dot(h_t_1, u_c) + b_c)

            f_t = sigmoid( np.dot( x_t, w_f ) + np.dot( h_t_1, u_f ) + b_f )

            c_t = i_t * c_t_tilda + f_t * c_t_1
            o_t = sigmoid( np.dot( x_t, w_o ) + np.dot( h_t_1, u_o ) + b_o )
            h_t = o_t * softsign( c_t )

            h_t_1 = h_t
            c_t_1 = c_t
        
        self.h_t_1, self.c_t_1 = h_t_1, c_t_1
    
    
    def pred(self, x, iterations = 1):
        h_t_1 , c_t_1 = self.h_t_1, self.c_t_1
        x_t = x.copy()
        l_x = []
        for _ in range(iterations):
            x_t = x_t.reshape(1,1)
            i_t = sigmoid( np.dot(x_t, w_i) + np.dot(h_t_1, u_i) + b_i )
    
            c_t_tilda = softsign(np.dot(x_t, w_c) + np.dot(h_t_1, u_c) + b_c)

            f_t = sigmoid( np.dot( x_t, w_f ) + np.dot( h_t_1, u_f ) + b_f )

            c_t = i_t * c_t_tilda + f_t * c_t_1
            o_t = sigmoid( np.dot( x_t, w_o ) + np.dot( h_t_1, u_o ) + b_o )
            h_t = o_t * softsign( c_t )

            h_t_1 = h_t
            c_t_1 = c_t
        
            x_t = np.dot( h_t , w_out ) + b_out
            l_x.append(x_t.item())
        
        self.h_t_1 = h_t_1
        self.c_t_1 = c_t_1
        return np.array(l_x)
```

## Functions Regex

```python
import re

def get_act_fnc(l_a):
    string = ''
    for activation in set(l_a):
        pattern = activation + r'.*'
        result = re.search(pattern,s_code_act)
        if result:
            string += result.group() + '\n'
            
    string += '\nd_act = { ' + ', '.join([ f"'{act}' : {act}" for act in set(l_a)]) + ' }\n\n'
    return string

print(get_act_fnc(l_a))
```
